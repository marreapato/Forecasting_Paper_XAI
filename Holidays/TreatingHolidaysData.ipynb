```python
import pandas as pd
sheet_id = "19Qm-gxnILXY8QTBASGfwfxp6NsLMOLuEIVmJbQJF8r4"
sheet_name = "holidays_all_countries"
url = f"https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}"
```

```python
feriados = pd.read_csv(url,header = None)
feriados.head()
```

```python
feriados.loc[feriados[4]=='observance',3].unique()
```

```python
feriados.loc[feriados[4]=='optional',3].unique()
```

```python
feriados.loc[feriados[4]=='public',3].unique()
```

```python
df3 = pd.get_dummies(feriados,
                     columns = [4])

display(df3)
```

```python
feriadosfim=df3[[0,2,"4_observance","4_optional","4_public"]]
feriadosfim = feriadosfim.rename(columns={0:"Datestart",2:"Dateend"})
feriadosfim["Datestart"] = pd.to_datetime(feriadosfim["Datestart"]).dt.date
feriadosfim["Dateend"] = pd.to_datetime(feriadosfim["Dateend"]).dt.date
feriadosfim.head()
```

```python
df = pd.DataFrame(feriadosfim)
df['Datestart'] = pd.to_datetime(df['Datestart'])
df['Dateend'] = pd.to_datetime(df['Dateend'])

# Function to expand date range for each holiday
def expand_dates(row):
    date_range = pd.date_range(row['Datestart'], row['Dateend'], freq='D')
    return pd.Series(date_range, name='Date')

# Apply the function to each row and concatenate the results
expanded_df = df.apply(expand_dates, axis=1).stack().reset_index(level=1, drop=True).reset_index()
expanded_df
```

```python
df['index'] = df.index
df
```

```python
expanded_df.columns = ['index','Date']

# Merge with the original DataFrame to include other columns
result_df = pd.merge(expanded_df, df, left_on=['index'], right_index=True, how='left')
result_df
```

```python
# Fill NaN values with 0 for non-holiday dates
result_df = result_df[['Date','4_observance','4_optional','4_public']]

result_df
```
